```python
import csv
import numpy as np
import pandas
import matplotlib.pyplot as plt

log_data = []

with open(f"data/performance_test_3obj_cpu.csv") as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        log_data.append(row)
        
log_data = np.array(log_data, dtype=np.float32)
print(pandas.DataFrame(log_data))
```

```python
log_data_2 = []

with open(f"data/performance_test_3obj_cpu_grid.csv") as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        log_data_2.append(row)
        
log_data_2 = np.array(log_data_2, dtype=np.float32)
print(pandas.DataFrame(log_data_2))
```

```python
print(pandas.DataFrame(log_data_2[:, 1] / log_data[:, 1], log_data_2[:, 0].astype('int')))
```

```python
plt.rcParams['figure.figsize'] = [16, 10]
plt.rcParams['font.size'] = '14'

data = log_data
data2 = log_data_2

x1 = data[:, 0]
y1 = data[:, 1]
y2 = data2[:, 1]

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x1, y1, linestyle='-', marker='s', color='indigo', linewidth=4, label="Jednowątkowe CPU bez siatki sąsiedztwa (a)")
ax.plot(x1, y2, linestyle='-', marker='s', color='orange', linewidth=4, label="Jednowątkowe CPU z siatką sąsiedztwa (b)")

font_size = 20
# ax.set_title("Wydajność w zależności od długości boku dwóch obiektów", fontsize=font_size)
ax.set_ylabel("Iteracje na sekundę", fontsize=font_size)
ax.set_xlabel("Liczba węzłów", fontsize=font_size)
ax.legend(loc="upper right", fontsize=font_size)
ax.set_yscale("log")

for i, j in zip(x1, y1):
    i = int(i)
    j = int(j)
    if i > 700:
        ax.annotate(str(j), xy=(i,j), xytext=(2,8), textcoords='offset points')

for i, j in zip(x1, y2):
    i = int(i)
    j = int(j)
    if i > 700:
        ax.annotate(str(j), xy=(i,j), xytext=(2,8), textcoords='offset points')

plt.show()
```
